### 오픈소스 형식 형태소 분석기 5가지

- 한나눔은 너무 성능이 떨어져서 제외. 일단표기는 해둠
- 카카오 형태소 분석기 카이를 제외한 이유는 사전 추가가 안된다.

- 트위터 분석기, 코모란 두 가지를 중점

```python
okt = Okt()
han = Hannanum()
komoran = Komoran()
ko = Kkma()
```

```python
# 데이터 입수
text = open( '타이레놀.txt' ).read() 
```

#### 트위터 부분

```python
# 트위터 morphs
okt_result = okt.morphs(text)
# 트위터 nouns 
okt_result1 = okt.nouns(text)
# 트위터 phrases
okt_result2 = okt.phrases(text)
# 트위터 pos 
okt_result3 = okt.pos(text)
```

```python
# 트위터 분석기 
print(okt_result3)
```

### 코모란 부분

```python
komoran_result = komoran.nouns("\n".join([x for x in text.split("\n") if x]))
komoran_result1 = komoran.morphs("\n".join([x for x in text.split("\n") if x]))
komoran_result2 = komoran.pos("\n".join([x for x in text.split("\n") if x]))
```

```python
# 코모란 
komoran_result2
```
